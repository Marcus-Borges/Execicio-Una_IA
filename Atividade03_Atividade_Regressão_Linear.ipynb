```python
# Importando as bibliotecas necessárias para manipulação de dados, modelagem de regressão linear e visualização.
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Criando uma estrutura de dados usando Pandas com os dados fornecidos.
data = {
    "Ano": [2023, 2022, 2021, 2020, 2019, 2018, 2016, 2015, 2014, 2013, 2023],
    "Numero_Mulheres_TI": [34, 28, 25, 16, 35, 27, 26, 8, 11, 15, 34]
}
df = pd.DataFrame(data)

# Exibindo o resumo estatístico dos dados
resumo_estatistico = df.describe()
print(resumo_estatistico)

# Removendo duplicatas e ordenando por ano para tornar o conjunto de dados consistente para análise.
df = df.drop_duplicates().sort_values("Ano").reset_index(drop=True)

# Reestruturando os dados para ajuste do modelo.
X = df[['Ano']].values
y = df['Numero_Mulheres_TI'].values

# Criando e ajustando um modelo de Regressão Linear.
model = LinearRegression()
model.fit(X, y)

# Prevendo o número de mulheres em TI para o ano de 2024.
predicted_2024 = model.predict([[2024]])[0]

# Gerando previsões ao longo dos anos no conjunto de dados para plotagem.
years = np.arange(2013, 2025).reshape(-1, 1)
predictions = model.predict(years)

# Plotando os dados e a linha de regressão.
plt.figure(figsize=(10, 6))
plt.scatter(df['Ano'], df['Numero_Mulheres_TI'], color="blue", label="Dados Reais")
plt.plot(years, predictions, color="red", label="Reta de Regressão Linear")
plt.scatter([2024], [predicted_2024], color="green", marker="x", s=100, label=f"Previsão para 2024: {int(predicted_2024)}")

# Rotulando os eixos e adicionando uma legenda.
plt.xlabel("Ano")
plt.ylabel("Número de Mulheres na Área de TI")
plt.title("Regressão Linear do Número de Mulheres na Área de TI ao longo dos anos")
plt.legend()
plt.grid(True)
plt.show()

predicted_2024
```
