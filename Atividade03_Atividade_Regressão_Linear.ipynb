```python
# Importando as bibliotecas necessrias para manipulao de dados, modelagem de regresso linear e visualizao.
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Criando uma estrutura de dados usando Pandas com os dados fornecidos.
data = {
    "Ano": [2023, 2022, 2021, 2020, 2019, 2018, 2016, 2015, 2014, 2013, 2023],
    "Numero_Mulheres_TI": [34, 28, 25, 16, 35, 27, 26, 8, 11, 15, 34]
}
df = pd.DataFrame(data)

# Exibindo o resumo estatstico dos dados
resumo_estatistico = df.describe()
print(resumo_estatistico)

# Removendo duplicatas e ordenando por ano para tornar o conjunto de dados consistente para anlise.
df = df.drop_duplicates().sort_values("Ano").reset_index(drop=True)

# Reestruturando os dados para ajuste do modelo.
X = df[['Ano']].values
y = df['Numero_Mulheres_TI'].values

# Criando e ajustando um modelo de Regresso Linear.
model = LinearRegression()
model.fit(X, y)

# Prevendo o nmero de mulheres em TI para o ano de 2024.
predicted_2024 = model.predict([[2024]])[0]

# Gerando previses ao longo dos anos no conjunto de dados para plotagem.
years = np.arange(2013, 2025).reshape(-1, 1)
predictions = model.predict(years)

# Plotando os dados e a linha de regresso.
plt.figure(figsize=(10, 6))
plt.scatter(df['Ano'], df['Numero_Mulheres_TI'], color="blue", label="Dados Reais")
plt.plot(years, predictions, color="red", label="Reta de Regresso Linear")
plt.scatter([2024], [predicted_2024], color="green", marker="x", s=100, label=f"Previso para 2024: {int(predicted_2024)}")

# Rotulando os eixos e adicionando uma legenda.
plt.xlabel("Ano")
plt.ylabel("Nmero de Mulheres na rea de TI")
plt.title("Regresso Linear do Nmero de Mulheres na rea de TI ao longo dos anos")
plt.legend()
plt.grid(True)
plt.show()

predicted_2024
```
